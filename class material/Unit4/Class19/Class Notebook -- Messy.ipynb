```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('housing.csv')
```

```python
X = df.iloc[:, :-1]
y = df['PRICE']
```

```python
lreg = LinearRegression()
```

```python
lreg.fit(X, y)
```

```python
lreg.coef_
```

```python
lreg.intercept_
```

```python
np.sum(X*lreg.coef_, axis=1) + lreg.intercept_
```

```python
X.dot(lreg.coef_)
```

```python
lreg.coef_ = np.hstack([lreg.coef_, lreg.coef_]).reshape(13, 2)
```

```python
lreg.coef_
```

```python
coeff     = np.random.normal(0, 0.1, size=39).reshape(13, 3)
intercept = np.random.normal(0, 0.1, size=3)
```

```python
X_hidden = X.dot(coeff) + intercept
```

```python
X_hidden = np.maximum(0, X_hidden)
```

```python
X_hidden2 = np.maximum(0, X_hidden2)
```
